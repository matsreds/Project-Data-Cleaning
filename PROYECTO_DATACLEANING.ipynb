```python
def cambiando_nombres(steam_df):
    """
    Esta funcion renombras las columnas
    del DataFrame

      """

    df_1 = steam_df.copy() #siempre hacer una copia del df cuando este trabajando con un funcion
    df_1.columns = ['URL', 'Tipos', 'Nombre', 'Descripcion', 'Reseñas_recientes', 'Todas_las_reseñas','Fecha_lanzamiento', 'Diseñador',
              'Editor', 'Etiquetas_populares', 'Detalles_juego', 'Lenguajes', 'Logros', 'Genero','Descripcion_juego', 'Contenido_maduro',
              'Requisitos_minimos', 'Requisitos_recomendados', 'Precio_original', 'Precio_descuento']

    return df_1
```

```python
def borrando_columnas(steam_df):
    """
       Esta funcion borra las columnas que no necesitamos

    """

    df_1 = steam_df.copy() #siempre hacer una copia del df cuando este trabajando con un funcion
    df_1.columns
    steam_to_drop = ['Logros','Reseñas_recientes','Contenido_maduro','Precio_descuento','Etiquetas_populares']
            
    for columna in steam_to_drop:
                     steam_df[columna].drop(axis=1, inplace=True)
    


    return df_1
```

```python
def acomodando_texto(steam_df):
    """
    Esta funcion elimna espacios en los nombres de las columnas
    además cambia la primer letra por mayuscula y las demás en minusculas

    """
    
    df_1 = steam_df.copy()
    df_1['Descripcion_juego'] = df_1['Descripcion_juego'].str.lstrip()
    
    steam_to_capitalize = ['Tipos', 'Descripcion', 'Descripcion_juego']
    for columna in steam_to_capitalize:
        steam_df[columna].str.capitalize()
        
        
    return df_1 
    
    
```

```python
def modificando_NaN(steam_df):
    """
       Esta funcion modifica los valores NaN 
       por 'Nos specified'

    """
    df_1 = steam_df.copy()
    
    to_fillna = ['Requisitos_recomendados','Requisitos_minimos','Descripcion', 'Todas_las_reseñas', 'Editor', 'Detalles_juego','Descripcion_juego',
            'Genero','Fecha_lanzamiento','Precio_original']
    for columna in to_fillna:
        df_1[columna].fillna('Not specified', inplace=True)
        
    return df_1
    
```

```python
def eliminando_registros(steam_df):
    """
        Esta funcion elimina los registros de la columna nombre
        que tienen valor NaN

    """
    
    df_1 = steam_df.copy()
    nombre = df_1[df_1['Nombre'] == 'Not specified']
    borrando_registros = df_1.drop[nombre]
    
    return df_1
```

```python
cambio_nombres = cambiando_nombres(steam_df)
columnas_eliminadas = borrando_columnas(steam_df)
texto_acomodado = acomodando_texto(steam_df)
NaN_modificados = modificando_NaN(steam_df)
registros_eliminados = eliminando_registros(steam_df)
```
